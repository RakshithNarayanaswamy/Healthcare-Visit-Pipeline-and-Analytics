BRONZE TABLES

```python
%sql
-- SQL
USE CATALOG hospital_cg;
USE SCHEMA `visit-check`;
```

```python
import dlt 
from pyspark.sql.functions import * 
from pyspark.sql.types import *
```

```python
BASE = "/Volumes/hospital_cg/visit-check/raw"

@dlt.table(name="bronze_patients")
def bronze_patients():
    return (spark.readStream
            .format("cloudFiles")
            .option("cloudFiles.format","csv")
            .option("cloudFiles.includeExistingFiles","true")
            .option("header","true").option("multiLine","true")
            .load(f"{BASE}/patients"))         

@dlt.table(name="bronze_providers")
def bronze_providers():
    return (spark.readStream
            .format("cloudFiles").option("cloudFiles.format","csv")
            .option("cloudFiles.includeExistingFiles","true")
            .option("header","true").option("multiLine","true")
            .load(f"{BASE}/providers"))

@dlt.table(name="bronze_encounters")
def bronze_encounters():
    return (spark.readStream
            .format("cloudFiles").option("cloudFiles.format","csv")
            .option("cloudFiles.includeExistingFiles","true")
            .option("header","true").option("multiLine","true")
            .load(f"{BASE}/encounters"))

@dlt.table(name="bronze_conditions")
def bronze_conditions():
    return (spark.readStream
            .format("cloudFiles").option("cloudFiles.format","csv")
            .option("cloudFiles.includeExistingFiles","true")
            .option("header","true").option("multiLine","true")
            .load(f"{BASE}/conditions"))
```
